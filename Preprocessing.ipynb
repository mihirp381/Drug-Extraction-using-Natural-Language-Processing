```python
import ast
import pandas as pd

def read_list_of_dicts_from_file(file_path):
    list_of_dicts = []
    with open(file_path, 'r') as file:
        for line in file:
            # Use ast.literal_eval to safely evaluate the string as a Python dictionary
            # This assumes that each line in the file is a dictionary-like string
            try:
                dict_obj = ast.literal_eval(line.strip())
                list_of_dicts.append(dict_obj)
            except (SyntaxError, ValueError) as e:
                # Handle the error
                print(f"Error parsing line: {line.strip()}")
                print(f"Error message: {e}")
    return list_of_dicts

# Usage
file_path = '/code/Final_Run.txt'  # Replace with your actual file path
list_of_dicts = read_list_of_dicts_from_file(file_path)
drugs = []
for i in list_of_dicts:
    if i is not None: 
        for j in i:
            drugs.append(j)

df_pubmed = pd.DataFrame.from_records(drugs)

#Dataframe of all drugs extracted from PubMed Sources
```

```python
def extract_protein(x):
    
    doc = nlp(x[1])
    bcpr_keywords = ["bcrp", "breast cancer resistance protein", "abcg2"]
    
    for token in doc:
        if token.text.lower() in bcpr_keywords :
            return token.text

df_pubmed['Protein']= df_pubmed.apply(lambda x: extract_protein(x), axis = 1)
```

```python
df_biorxiv['Protein']= df_biorxiv.apply(lambda x: extract_protein(x), axis = 1)
```

```python
import re

# Function to tokenize and perform positional encoding on the context
def mark_entities(context, drug, protein):
    # Escaping the entities to avoid issues with special characters in regular expressions
    drug_escaped = re.escape(drug.lower())
    protein_escaped = re.escape(protein.lower())
    
    # Replace the drug and protein names with the marked entities
    # The word boundaries (\b) are used to ensure that only whole words are matched
    context_marked = re.sub(rf'\b{drug_escaped}\b', '<e1>' + drug.lower() + '</e1>', context.lower())
    context_marked = re.sub(rf'\b{protein_escaped}\b', '<e2>' + protein.lower() + '</e2>', context_marked.lower())
    
    return context_marked.lower()

# Apply the function to the dataset (for demonstration, let's apply it to the first 5 rows)
df_pubmed['context_marked'] = df_pubmed.apply(lambda row: mark_entities(row['context'], row['drug'], row['Protein']), axis=1)
```

```python
df_biorxiv['context_marked'] = df_biorxiv.apply(lambda row: mark_entities(row['context'], row['drug'], row['Protein']), axis=1)
```

```python
import requests
import time

def get_smiles_from_drug_name(drug_name, retries=3, backoff_factor=0.1):
    # Endpoint for searching compounds in PubChem
    search_url = f'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/{drug_name}/cids'
    for _ in range(retries):
        response = requests.get(search_url)
        print(response.status_code)
        if response.status_code == 200:
            data = response.json()
            cid = data['IdentifierList']['CID'][0]
            
            properties_url = f'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{cid}/property/CanonicalSMILES/JSON'
            response = requests.get(properties_url)
            
            if response.status_code == 200:
                properties_data = response.json()
                smiles = properties_data['PropertyTable']['Properties'][0]['CanonicalSMILES']
                return smiles
            else:
                raise Exception(f"Could not get properties for CID {cid}. Status code: {response.status_code}")
        elif response.status_code == 503:
            time.sleep((2 ** _) * backoff_factor)
        elif response.status_code == 404:
            pass  # No CID found for the drug name
        else:
            raise Exception(f"Search request failed for drug '{drug_name}'. Status code: {response.status_code}")

    raise Exception(f"All retries failed for drug '{drug_name}'.")

# Example usage:
smiles_string = get_smiles_from_drug_name('aspirin')
print(smiles_string)
```

```python
import requests
import time

def get_smiles_from_drug_name(drug_name, retries=3):
    # Endpoint for searching compounds in PubChem
    search_url = f'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/{drug_name}/cids/JSON'

    for attempt in range(retries):
        try:
            response = requests.get(search_url)
            response.raise_for_status()  # Will raise an exception for status codes 4xx/5xx
        except requests.exceptions.HTTPError as http_err:
            if response.status_code == 503:
                # Check if the response has a 'Retry-After' header
                retry_after = response.headers.get('Retry-After')
                wait_time = int(retry_after) if retry_after else (2 ** attempt)
                print(f"503 Error, retrying after {wait_time} seconds.")
                time.sleep(wait_time)
            else:
                print(f"HTTP error occurred: {http_err}")
        except requests.exceptions.RequestException as err:
            print(f"Other error occurred: {err}")
        else:
            # No exceptions, successful request
            data = response.json()
            cid = data['IdentifierList']['CID'][0]
            
            properties_url = f'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{cid}/property/CanonicalSMILES/JSON'
            response = requests.get(properties_url)
            if response.status_code == 200:
                properties_data = response.json()
                smiles = properties_data['PropertyTable']['Properties'][0]['CanonicalSMILES']
                return smiles
            else:
                print(f"Could not get properties for CID {cid}. Status code: {response.status_code}")
                return None

    print("Failed to retrieve data after retries.")
    return None

# Example usage:
smiles_string = get_smiles_from_drug_name('apixaban')
print(smiles_string if smiles_string else "No SMILES string retrieved.")
```
